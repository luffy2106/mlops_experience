# Question answering overview

Question answering tasks return an answer given a question. If you've ever asked a virtual assistant like Alexa, Siri or Google what the weather is, then you've used a question answering model before. There are two common types of question answering tasks:

- Extractive: extract the answer from the given context.
- Abstractive: generate an answer from the context that correctly answers the question.

This guide will show you how to:

1. Finetune [DistilBERT](https://huggingface.co/distilbert-base-uncased) on the [SQuAD](https://huggingface.co/datasets/squad) dataset for extractive question answering.
2. Use your finetuned model for inference.

## Preprocess

The next step is to load a DistilBERT tokenizer to process the `question` and `context` fields:

````python
from transformers import AutoTokenizer

# tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")
````

There are a few preprocessing steps particular to question answering tasks you should be aware of:

1. Some examples in a dataset may have a very long `context` that exceeds the maximum input length of the model. To deal with longer sequences, truncate only the `context` by setting `truncation="only_second"`.
2. Next, map the start and end positions of the answer to the original `context` by setting
   `return_offset_mapping=True`.
3. With the mapping in hand, now you can find the start and end tokens of the answer. Use the [sequence_ids](https://huggingface.co/docs/tokenizers/main/en/api/encoding#tokenizers.Encoding.sequence_ids) method to
   find which part of the offset corresponds to the `question` and which corresponds to the `context`.

Here is how you can create a function to truncate and map the start and end tokens of the `answer` to the `context`:

````python
def preprocess_function(examples):
    questions = [q.strip() for q in examples["question"]]
    inputs = tokenizer(
        questions,
        examples["context"],
        max_length=384,
        truncation="only_second",
        return_offsets_mapping=True,
        padding="max_length",
    )

    offset_mapping = inputs.pop("offset_mapping")
    answers = examples["answers"]
    start_positions = []
    end_positions = []

    for i, offset in enumerate(offset_mapping):
        answer = answers[i]
        start_char = answer["answer_start"][0]
        end_char = answer["answer_start"][0] + len(answer["text"][0])
        sequence_ids = inputs.sequence_ids(i)

        # Find the start and end of the context
        idx = 0
        while sequence_ids[idx] != 1:
            idx += 1
        context_start = idx
        while sequence_ids[idx] == 1:
            idx += 1
        context_end = idx - 1

        # If the answer is not fully inside the context, label it (0, 0)
        if offset[context_start][0] > end_char or offset[context_end][1] < start_char:
            start_positions.append(0)
            end_positions.append(0)
        else:
            # Otherwise it's the start and end token positions
            idx = context_start
            while idx <= context_end and offset[idx][0] <= start_char:
                idx += 1
            start_positions.append(idx - 1)

            idx = context_end
            while idx >= context_start and offset[idx][1] >= end_char:
                idx -= 1
            end_positions.append(idx + 1)

    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions
    return inputs
````

To apply the preprocessing function over the entire dataset, use ü§ó Datasets [map](https://huggingface.co/docs/datasets/main/en/package_reference/main_classes#datasets.Dataset.map) function. You can speed up the `map` function by setting `batched=True` to process multiple elements of the dataset at once. Remove any columns you don't need:

````python
tokenized_data_folder = "../data/preprocessed"


# Check if the tokenized dataset already exists
if os.path.exists(tokenized_data_folder):
    # Load the tokenized data if it already exists
    tokenized_squad = load_from_disk(tokenized_data_folder)
    print("Loaded tokenized data from disk.")
else:
    # If the tokenized data doesn't exist, preprocess and save it
    print("Tokenized data not found. Tokenizing and saving to disk...")
    # Apply the preprocessing function to tokenize the dataset
    tokenized_squad = squad.map(preprocess_function, batched=True, remove_columns=squad["train"].column_names)
    
    # Save the tokenized data to the specified directory in Arrow format
    tokenized_squad.save_to_disk(tokenized_data_folder)
    print("Tokenized data saved to disk.")




# tokenized_squad = squad.map(preprocess_function, batched=True, remove_columns=squad["train"].column_names)
````

#### 2. Log Model Performance Metrics (Accuracy, F1, Perplexity)
Define a compute_metrics function and pass it to the Trainer. This will calculate and log evaluation metrics such as accuracy, F1, or perplexity during each evaluation step.

````python
from sklearn.metrics import accuracy_score, f1_score

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)  # Assuming classification
    acc = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds, average="weighted")  # Use "macro" or "micro" as needed
    return {"accuracy": acc, "f1": f1}
````

#### 3. Log Predictions, Ground Truth Labels, and Samples for Analysis
During each evaluation, you can log sample inputs, predictions, and ground truth labels to MLflow as artifacts. Save these as text or JSON files and upload them to MLflow

````python
import json
from mlflow import log_artifact

def log_predictions(trainer, dataset, artifact_dir, split_name="eval"):
    predictions, labels, _ = trainer.predict(dataset)
    pred_labels = predictions.argmax(axis=-1)  # Convert to label format if necessary

    # Prepare log data
    samples = [{"prediction": int(pred), "label": int(label)} for pred, label in zip(pred_labels, labels)]
    
    # Save and log as artifact

    artifact_file = artifact_dir + f"/{split_name}_predictions.json" 
    with open(artifact_file, "w") as f:
        json.dump(samples, f)
    log_artifact(artifact_file)
````

#### 4. Track System Performance (Memory, CPU/GPU Utilization)
MLflow itself doesn‚Äôt track system metrics directly, but you can use system monitoring tools like psutil (for CPU and memory usage) and GPUtil (for GPU usage) and log these metrics to MLflow.

````python
import psutil
import GPUtil
import time

def log_system_metrics(step):
    # Log CPU and memory usage
    cpu_usage = psutil.cpu_percent()
    memory_usage = psutil.virtual_memory().percent
    mlflow.log_metric("cpu_usage", cpu_usage, step=step)
    mlflow.log_metric("memory_usage", memory_usage, step=step)
    
    # Log GPU usage (if GPU is available)
    gpus = GPUtil.getGPUs()
    if gpus:
        gpu = gpus[0]  # Assumes using the first GPU
        mlflow.log_metric("gpu_usage", gpu.load * 100, step=step)
        mlflow.log_metric("gpu_memory", gpu.memoryUtil * 100, step=step)

# Use this function in a loop or callback to log periodically during training
````

#### 5. Track Parameter and Hyperparameter Values Over Time (e.g., Learning Rate)
For tracking parameters and hyperparameters over time, such as the learning rate, log them manually using mlflow.log_metric() in a callback function or training loop.

````python
from transformers import TrainerCallback
# Define callback to log learning rate over time
class LogLRCallback(TrainerCallback):
    def on_step_begin(self, args, state, control, **kwargs):
        current_lr = args.learning_rate
        mlflow.log_metric("learning_rate", current_lr, step=state.global_step)

# trainer.add_callback(LogLRCallback())
````

````python
import mlflow
from transformers import AutoModelForQuestionAnswering, TrainingArguments, Trainer

mlflow.set_tracking_uri("../models/mlruns")
# Set experiment name explicitly
mlflow.set_experiment("flan-t5-qa-experiment")


# End any existing run if active
if mlflow.active_run():
    mlflow.end_run()

# Set up MLflow
mlflow.start_run(run_name="flan-t5-qa")





model = AutoModelForQuestionAnswering.from_pretrained("google/flan-t5-small")

training_args = TrainingArguments(
    output_dir="../models/checkpoints",
    evaluation_strategy="steps",
    learning_rate=2e-5,
    eval_steps=10,                      # Log training loss every 10 steps  
    logging_steps=10,                   # Log evaluation loss Evaluation every 10 steps
    gradient_accumulation_steps = 2,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    num_train_epochs=3,
    weight_decay=0.01,
    push_to_hub=False,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_squad["train"],
    eval_dataset=tokenized_squad["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)
# Log system performance periodically
for step in range(0, len(trainer.get_train_dataloader())):
    log_system_metrics(step)

trainer.train()


# Log model to MLflow
mlflow.log_param("model", "flan-t5")
mlflow.log_metric("epoch", training_args.num_train_epochs)
mlflow.transformers.log_model(model, "flan-t5-qa-model")

# Log predictions and metrics as artifacts
log_predictions(trainer, tokenized_squad["test"], "test")

mlflow.end_run()
````

## Inference

Great, now that you've finetuned a model, you can use it for inference!

Come up with a question and some context you'd like the model to predict:

````python
question = "How many programming languages does BLOOM support?"
context = "BLOOM has 176 billion parameters and can generate text in 46 languages natural languages and 13 programming languages."
````

The simplest way to try out your finetuned model for inference is to use it in a [pipeline()](https://huggingface.co/docs/transformers/main/en/main_classes/pipelines#transformers.pipeline). Instantiate a `pipeline` for question answering with your model, and pass your text to it:

````python
from transformers import pipeline

question_answerer = pipeline("question-answering", model="my_awesome_qa_model")
question_answerer(question=question, context=context)
````

You can also manually replicate the results of the `pipeline` if you'd like:

Tokenize the text and return PyTorch tensors:

````python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("my_awesome_qa_model")
inputs = tokenizer(question, context, return_tensors="pt")
````

Pass your inputs to the model and return the `logits`:

````python
import torch
from transformers import AutoModelForQuestionAnswering

model = AutoModelForQuestionAnswering.from_pretrained("my_awesome_qa_model")
with torch.no_grad():
    outputs = model(**inputs)
````

Get the highest probability from the model output for the start and end positions:

````python
answer_start_index = outputs.start_logits.argmax()
answer_end_index = outputs.end_logits.argmax()
````

Decode the predicted tokens to get the answer:

````python
predict_answer_tokens = inputs.input_ids[0, answer_start_index : answer_end_index + 1]
tokenizer.decode(predict_answer_tokens)
````
